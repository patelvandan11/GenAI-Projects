```python
import pandas as pd
import json

chunk_size = 10000  # Process 10,000 rows at a time
input_path = "C:/Users/Vandan Patel/Downloads/archive (1)/RecipeNLG_dataset.csv"
output_path = "recipes.json"

with open(output_path, "w", encoding="utf-8") as f:
    pass

for chunk in pd.read_csv(input_path, chunksize=chunk_size):
    with open(output_path, "a", encoding="utf-8") as f_out:
        for _, row in chunk.iterrows():
            record = {
                "instruction": "Generate a recipe title and step-by-step directions based only on the provided ingredients.",
                "input": row["ingredients"],
                "output": f"Title: {row['title']}\n\nDirections:\n{row['directions']}"
            }
            f_out.write(json.dumps(record) + "\n")
```
